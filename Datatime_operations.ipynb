### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data = pd.read_csv("C:/Data Sets/Data Sets/wind.data")

#data.dtypes
#Yr, Mo, Dy are int64

#d2 = data.loc[:,['Yr','Mo','Dy']]

#data.loc[:, data.dtypes=='int64'].applymap(lambda x: pd.to_datetime(x,format = '%Y'))
#pd.to_datetime(d2, format = '%d/%m/%Y')
#data.loc[:, ['Yr','Mo','Dy']]

data['date']=data['Yr'].astype('str')+data['Mo'].astype('str')+data['Dy'].astype('str')
data.drop(columns=['Yr','Mo','Dy'], inplace = True)
data['date']= pd.to_datetime(data.date, format = '%y%m%d')
data
```

### Step 4. Year 2061 is seemingly imporoper. Convert every year which are < 70 to 19XX instead of 20XX.

```python
np.where(data.date.dt.year>2020, data.date+pd.DateOffset(years=-100), data.date)
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data['date'] = np.where(data.date.dt.year>2020, data.date+pd.DateOffset(years=-100), data.date)
data.set_index('date', inplace = True)
```
